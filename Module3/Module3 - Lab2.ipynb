Create a 2d scatter plot that graphs the `area` and `perimeter` features:

```python
def plotGraphs(x, y, m):
    plt.scatter(df[x], df[y], alpha=0.75, marker=m)
    plt.title(x.title() + '  x  ' + y.title())
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
values = [['area', 'perimeter', '^'], ['groove', 'asymmetry', '.'], ['compactness', 'width', 'o']]
for i in values:
    plotGraphs(i[0], i[1], i[2])
```

Create a 2d scatter plot that graphs the `groove` and `asymmetry` features:

```python
plotGraphs('groove', 'asymmetry', '.')
```

Create a 2d scatter plot that graphs the `compactness` and `width` features:

```python
plotGraphs('compactness', 'width', 'o')
```
